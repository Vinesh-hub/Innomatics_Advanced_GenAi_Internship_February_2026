# Problem 1: Smart Light Controller

```python
# class 
class Smartlight:
    def __init__(self,name,action):
        self.name=name
        self.action=action
    def per_action(self):
        print(f"{self.name} is {self.action}")
# object creation
light=Smartlight("Bedroom Light","ON")
light.per_action()
        
        
```

# Problem 2: Employee ID Card System

```python
# class
class Employee:
    def __init__(self,E_name,E_id,dept):
                 self.E_name=E_name
                 self.E_id=E_id
                 self.dept=dept
    def display(self):
        print("Employee ID Card")
        print("Name:",self.E_name)
        print("ID:",self.E_id)
        print("Department:",self.dept)

# create object
emp=Employee("Vinesh","EMP102","IT")
emp.display()
    
```

```python
#class
class Contact:
    def __init__(self,name,mobile):
        self.name=name
        self.mobile=mobile
    def display(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Mobile:",self.mobile)

# input from user
name=input("Enter name")
mobile=input("Enter mobile:")
con=Contact(name,mobile)
con.display()
```

# Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display(self):
        print("Name:",self.name)
        print(f"Price: ₹{self.price}")
# object creating
prod=Product("HeadPhones",2499)
prod.display()
```

# Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def display(self):
        print("Name:",self.name)
        print("Rating:",self.rating)
# object creating
mov=Movie("Inception",4.8)
mov.display()
```
