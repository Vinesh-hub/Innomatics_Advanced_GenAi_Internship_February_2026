# Problem Statement 1: Employee Performance Bonus Eligibility

```python
#input
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

# finding highest performance score
highest_score=max(employees.values())

# finding the highest performers
performers=[name for name,performance in employees.items() if performance==highest_score]

#print the result
print("Top Performers Eligible for Bonus:",",".join(performers),f"(Score:{highest_score})")
```

# Problem Statement 2: Search Query Keyword Analysis

# Problem Statement 3: Sensor Data Validation

```python
# input

sensor_readings=[3, 4, 7, 8, 10, 12, 5]

valid_readings=[]

for i in range(len(sensor_readings)):
    if sensor_readings[i]%2==0:
        valid_readings.append((i,sensor_readings[i]))
        
print("Valid Sensor Readings (Hour,Value):",valid_readings)
```

# Problem Statement 5: Sales Spike Detection

```python
# input
sales = [1200, 1500, 900, 2200, 1400, 3000]

# calculating avg sales
avg_sales=sum(sales)/len(sales)

threshold=avg_sales*1.3
for i in range(len(sales)):
    if sales[i]>threshold:
        print(f"Day {i+1}:{sales[i]}")
```

# Problem Statement 6: Duplicate User ID Detection

```python
#input
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

userid_count={}

# counting the each userid
for i in user_ids:
    userid_count[i]=userid_count.get(i,0)+1
    
# printing the user id appears more than once

for userid,count in userid_count.items():
    if count>1:
        print(f"{userid} --> {count} times")
```
