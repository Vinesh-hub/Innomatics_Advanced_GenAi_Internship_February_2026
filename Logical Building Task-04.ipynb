# Problem 1: Mobile Recharge Validation System

```python
# checks for valid recharge
def valid_recharge(amount):
    valid_plans=[199, 299, 399, 599]
    if amount<50:
        print("Enter amount must be >=50")
        return False
    if amount not in valid_plans:
        print("invalid plan! Available Plans[199, 299, 399, 599]")
        return False
    return True

#executes until recharge is succsseful
while True:
    amount=int(input("Enter Recharge Amount"))
    if valid_recharge(amount):
        print("Recharge Successful")
        break
    else:
        print("Please try again")
```

# Problem 2: Inventory Reorder Alert System

```python
#function for inventory alert reorder
def inventory_reorder(stock_list):
    for item,stock in stock_list.items():
        if stock<15:
            print(f"{item} --> Need to Order (Stock:{stock})")
        else:
            print(f"{item} --> Stock Ok (Stock:{stock})")

#input
stock_list={
    "books":14,
    "pens":20,
    "charts":50,
    "plank":12
}

# function Call
inventory_reorder(stock_list)
```

# Problem 3: Student Result Processing System

```python
#function return student result
def student_result(student_marks):
    total=0
    for marks in student_marks:
        total+=marks
    avg_marks=total/len(student_marks)
    print("Avg_marks:",avg_marks)
    if avg_marks>=50:
        print("Result:Pass")
    else:
        print("Result:Fail")

#input marks
student_marks=[45,40,50,57,60]

# function call
student_result(student_marks)
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
# fuction to calculate fair 
def fare_calculator(distance,peak):
    base_fare=50
    total=base_fare+(12*distance)
    if peak.lower() =="yes":
        total=total+(total*0.25)
    return total

while True:
    distance=int(input("Enter distance in km"))
    peak=input("Enter (Yes/No)")
    if distance<=0:
        print("Enter Valid Distance")
        continue
    total_fare=fare_calculator(distance,peak)
    print("Total Fare:",total_fare)
    break
    
```

# Problem 5: Employee Attendance Eligibility 

```python
def attendance_percentage(attendance_list):
    p=0
    for i in attendance_list:
        if i.lower()=="p":
            p+=1
    percentage=(p/len(attendance_list))*100
    return "Eligible" if percentage>=75 else "Not Eligible"

#input
attendance_list=["P","A","p","P"]

#functiopn call
result=attendance_percentage(attendance_list)
print(result)
```

# Problem 6: Password Strength Checker

```python
def password_checker(password):
    has_digit=False
    has_special=False
    if len(password)<8:
        return "Weak password must be greater than length 7"
    for i in password:
        if i.isdigit():
            has_digit=True
        if i in "@#$":
            has_special=True
    if has_digit and has_special:
        return "Strong password"
    else:
        return "Weak password include digits and special characters"
#input
password="vvgv#$%"

#function call
check=password_checker(password)
print(check)
```
