# Problem 1: Social Media – Post Engagement Analyzer

```python
# function to analyze engagement on social media
def engagement_analyzer(likes_list):
    total_likes=0
    for likes in likes_list:
        total_likes+=likes
    print("Total Likes:",total_likes)
    print("Post Status: Viral Post" if total_likes>=1000 else "Post Status: Normal Engagement")


#input
likes_lists=[100,200,300,400]
#function call
engagement_analyzer(likes_lists)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python
#function that analyzes seasonal rainfall data.
def rainfall_analyzer(required_level,rainfall_data):
    total=0
    for i in rainfall_data:
        total+=i
    avg_rainfall=total/len(rainfall_data)
    print("Average Rainfall:",avg_rainfall)
    print("Rainfall Status: Adequate Rainfall" if avg_rainfall>=required_level else "Rainfall Status: Inadequate Rainfall")

#input
required_level=int(input("Enter required rainfall level"))
rainfall_data=list(map(int,input().split()))

# function call
rainfall_analyzer(required_level,rainfall_data)
```

# Problem 4: Social Media – Duplicate Account Detection

```python
# function to detect duplicate usernames in a system
def duplicate_usernames(user_names):
    Status=False
    if len(user_names)>len(set(user_names)):
        Status=True
    if Status:
        print("Duplicates Accounts Found: Yes")
        print("Status: Flag Accounts")
    else:
        print("Duplicates Accounts Found: No")
        print("Status: Accounts are Unique")

#input
user_names=["user1","user2","user3","user1"]

#fuction call
duplicate_usernames(user_names)

    
```

# Problem 7: System Monitoring – Application Health Checker

# Problem 8: Banking – Daily Transaction Limit Checker

```python
# function to validate daily transaction amount.
def validate_transaction_limit(daily_transactions):
    print("Transaction Amount:",sum(daily_transactions))
    print("Transaction Status:",end=" ")
    if sum(daily_transactions)<=50000:
        print("Approved")
    else:
        print("Rejected")

#input
daily_transactions=[100,10000,30000,20000]

# function call
validate_transaction_limit(daily_transactions)
    
```

# Problem 9: E-Learning – Student Attendance Eligibility

```python
 # function to calculate attendance percentage and check exam eligibility.
def exam_eligibility(student_attendance):
    present_days=0
    for i in student_attendance:
        if i.upper()=="P":
            present_days+=1
    percentage=(present_days/len(student_attendance))*100
    print("Attendance Percentage:",percentage)
    if percentage>=75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

#input
student_attendance=["A","P","p","p","p","p","a","p"]

#function call
exam_eligibility(student_attendance)

    
```
