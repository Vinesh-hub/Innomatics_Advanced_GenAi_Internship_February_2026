# 2.Pass/Fail Analyzer

```python
marks=[45,78,90,33,60]
passed_students=0
failed_students=0
for i in marks:    #iterates throughout the marks list
    if i>=50:
        passed_students+=1
    else:
        failed_students+=1
print("Total Passed Students:",passed_students)
print("Total Failed Students:",failed_students)
        
```

# 3.Simple Data Cleaner

```python
names=["Alice","bob","CHARLIE"]
Cleaned_names=[]
for i in names:
    Cleaned_names.append(i.strip().lower())   #Strip removes the spaces(left,right) in a string,  Lower() converts string to lowercase
print(Cleaned_names)
```

# 4.Messege Length Analyzer 

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    print("Messege:",i)
    print("Length:",len(i))
    if len(i)>10:
        print("This is a long messege")
    print()
```
